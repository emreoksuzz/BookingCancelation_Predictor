```python
df["date of reservation"] = pd.to_datetime(df["date of reservation"], errors='coerce')
df['Month'] = df['date of reservation'].dt.month
df['Day'] = df['date of reservation'].dt.day
df = df[df["date of reservation"] != '2018-02-29'] #important feature 29th day need to be excluded 
```

```python
def categorize_session(row):
    if (row['Month'] >= 6 and row['Month'] <= 8) or (row['Month'] == 9 and row['Day'] <= 21):
        return 'Summer'
    elif (row['Month'] >= 3 and row['Month'] <= 5) or (row['Month'] == 6 and row['Day'] >= 21) or (row['Month'] == 9 and row['Day'] >= 22):
        return 'Spring'
    elif (row['Month'] >= 12) or (row['Month'] <= 2):
        return 'Winter'
    else:
        return 'Fall'
```

```python
df['Season'] = df.apply(categorize_session, axis=1)
```
